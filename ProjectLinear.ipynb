```python
import numpy as np
import pandas as pd

pd.set_option('max_columns', None)
pd.set_option('max_colwidth', None)

df_heart = pd.DataFrame()
df_heart = pd.read_csv('Heart.csv')
df_heart
```

```python
from sklearn.model_selection import train_test_split
feature_cols = ['Age','RestingBP','Cholesterol','FastingBS',
                'MaxHR','Oldpeak']

X = df_heart[feature_cols]
y = df_heart['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35,random_state=3)
                
                
```

```python
from sklearn.linear_model import LinearRegression
#creating new linear regression model
my_lin_reg = LinearRegression()
my_lin_reg.fit(X_train,y_train)

y_pred = my_lin_reg.predict(X_test)

print(y_pred)
```

```python
from sklearn import metrics
#calculating mse
mse = metrics.mean_squared_error(y_test, y_pred)
mse
```

```python
#calculating rmse(accuracy score)
rmse = np.sqrt(mse)
print('Below is our accuracy score before applying One Hot Coding:')
print(rmse)
print('Unlike other methods we use RMSE as our accuracy score. The lower the number for RMSE the better. \nThink of RMSE being the chances that the prediction is wrong, meaning having a .39 RMSE means that \nour accuracy score would be .59')
```

```python
print('We will now be applying One Hot Coding')
```

```python
#applying One Hot Coding
sex_dum = pd.get_dummies(df_heart.Sex)
cpt_dum = pd.get_dummies(df_heart.ChestPainType)
recg_dum = pd.get_dummies(df_heart.RestingECG)
ea_dum = pd.get_dummies(df_heart.ExerciseAngina)
sts_dum = pd.get_dummies(df_heart.ST_Slope)
OHC_df = pd.concat([df_heart, sex_dum, cpt_dum, recg_dum, ea_dum, sts_dum], axis='columns')
#new df with OHC applied
OHC_df
```

```python
#new feature columns to fit with One Hot Coding
new_feature_cols = ['Age','RestingBP','Cholesterol','FastingBS','MaxHR',
                    'Oldpeak','F','M','ASY','ATA','NAP','TA','LVH',
                    'Normal','ST','N','Y','Down','Flat','Up']
D = OHC_df[new_feature_cols]
e = OHC_df['HeartDisease']
#splitting the new data
D_train, D_test, e_train, e_test = train_test_split(D, e, test_size=0.35,random_state=3)
```

```python
#creating new linear regression model
linreg = LinearRegression()
linreg.fit(D_train,e_train)

y_prediction = linreg.predict(D_test)

print(y_prediction)
```

```python
#calculating mse
OHC_mse = metrics.mean_squared_error(e_test, y_prediction)
OHC_mse
```

```python
#Calculating rmse
OHC_rmse = np.sqrt(mse)
print('RMSE for data frame with One Hot Coding:')
print(OHC_rmse)
```

```python
print('As we can see the RMSE is the same for both before and after applying One Hot Coding for the linear regression model')
```
